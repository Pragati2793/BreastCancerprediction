```python
def models(X_train,Y_test):
    
    # LOGISTIC REGRESSION
    
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression()
    log.fit(X_train, Y_train)
    
    # Decision Tree
    
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier()
    tree.fit(X_train, Y_train)
    
    # Random Forest
    
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier()
    forest.fit(X_train, Y_train)
    
    # Print Accuracy of Training Data:
    print("1. Logistic Reg training Accuracy:",log.score(X_train,Y_train))
    print("2. Decision Tree training Accuracy:",tree.score(X_train,Y_train))
    print("3. Random Forest training Accuracy:",forest.score(X_train,Y_train))
    
    return log,tree,forest
```

### Model Training and Evaluation

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, Y_train)
y_pred_log=log.predict(X_test)
    
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)
y_pred_tree=tree.predict(X_test)
    
#Random Forest
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_train, Y_train)
y_pred_forest=forest.predict(X_test)



def classification_Evaluation(X_train,Y_train,X_test,Y_test,model):
    
    # Model Fitting
    model.fit(X_train, Y_train)
    
    # Model Predict
    
    y_pred_test=model.predict(X_test)
    
    print("Confusion matrix is: \n")
    # Ploting confusion Matrix:
    from sklearn import metrics
    metrics.plot_confusion_matrix(model,X_test,Y_test)
    plt.show()
    
    # Confusion Matrix
    cm=metrics.confusion_matrix(Y_test,y_pred_test)
    tp=cm[0][0]
    fp=cm[0][1]
    fn=cm[1][0]
    tn=cm[1][1]
    
    # Accuracy:
    Accuracy=(tp+tn)/(tp+tn+fp+fn)
    
    # recall:
    Recall=tp/(tp+fn)
    
    # precision:
    Precision=tp/(tp+fn)
    
    # F1_score:
    F1_score=(2*Precision*Recall)/(Precision+Recall)
    
    # Calculation Different Matrics
    
    fpr,tpr,threshold=metrics.roc_curve(Y_test,y_pred_test)
    
    #Auc    
    auc=metrics.roc_auc_score(Y_test,y_pred_test)
    
    fpr,tpr,threhsolds=metrics.roc_curve(Y_test,y_pred_test)
    
    print("*"*20)
    
    print("ROC PLotting is: \n")
    
    print("Classification Report is \n",metrics.classification_report(Y_test,y_pred_test))
    
    print("*"*20)
    # ROC PLotting
    plt.plot(fpr, tpr, color='orange', label='ROC curve (area = %0.3f)' % auc)
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--',label='ROC base model')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()
    
    # Making DataFrame
    
    #taking values in Dataframe
    values=np.array([[Accuracy],[Precision],[Recall],[F1_score],[auc]])
    index1=["Accuracy","Precision","Recall","F1_Score","AUC "]
    
    ans=pd.DataFrame(data=values,index=index1,columns=["Values"],dtype="float")  
    pd.set_option('display.float_format', '{:.4f}'.format)
    
    return ans
```

```python
# Logistic Regression
classification_Evaluation(X_train,Y_train,X_test,Y_test,log)
```
